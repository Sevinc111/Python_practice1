Create a `while` loop that prints all odd numbers from 0 to 30 on the same row.

> Hint: There are two ways to create the odd values!

```python
# Way 1: check for odd with modulo
i = 0
while i <= 30:
    if i % 2 != 0:
        print(i, end=' ')
    i += 1

# Way 2: start at 1, step by 2
i = 1
while i <= 30:
    print(i, end=' ')
    i += 2
```

## All In — Conditional Statements, Functions, and Loops

Use a `while` loop to count the number of values lower than 20 in the `nums` list.

```python
nums = [1, 12, 24, 31, 51, 70, 100]

count = 0
item = 0
while item < len(nums):
    if nums[item] < 20:
        count += 1
    item += 1

print("Values lower than 20:", count)
```

## Iterating over Dictionaries

Calculate how much Jan spent on products with a price of **$5 or more**.

```python
prices = {
    "box_of_spaghetti": 4,
    "lasagna": 5,
    "hamburger": 2
}
quantity = {
    "box_of_spaghetti": 6,
    "lasagna": 10,
    "hamburger": 0
}

money_spent = 0

for product in prices:
    if prices[product] >= 5:
        money_spent += prices[product] * quantity[product]

print("Spent on products >= $5:", money_spent)
```

How much did Jan spend on products that cost **less than $5**?

```python
money_spent = 0

for product in prices:
    if prices[product] < 5:
        money_spent += prices[product] * quantity[product]

print("Spent on products < $5:", money_spent)
```
